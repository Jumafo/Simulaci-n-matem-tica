```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

n_steps =200
steps = np.random.choice([-1,1], size=n_steps)
price =100 + np.cumsum(steps)

plt.figure(figsize=(10,5))
plt.plot(price)
plt.title("Random Walk de un precio")
plt.xlabel("Tiempo")
plt.ylabel("Precio")
plt.show()
```

**- Muchas simulaciones: aparece la distribucin normal**

```python
np.random.seed(1)

n_sim =5000
n_steps =200

final_positions = []

for i in range(n_sim):
    steps = np.random.choice([-1,1], size=n_steps)
    final_positions.append(np.sum(steps))

plt.hist(final_positions, bins=50)
plt.title("Distribucin de posiciones finales")
plt.show()

```

**- La incertidumbre crece con el tiempo**



```python
np.random.seed(0)

n_sim = 2000
times = [10, 50, 100, 200, 400]

for t in times:
    finals = []
    for i in range(n_sim):
        steps = np.random.choice([-1, 1], size=t)
        finals.append(np.sum(steps))
    print(f"t={t}, var{np.var(finals):.2f}")
```

**- Primera aproximacin a movimiento browniano**

```python
np.random.seed(42)

T =1
n =1000
dt = T/n

increments = np.random.normal(0, np.sqrt(dt), size=n)
W = np.cumsum(increments)

plt.plot(W)
plt.title("Aproximacin a movimiento browniano")
plt.show()

```

# Actividad

Simular 200 random walks y graficarlos juntos.

```python
plt.figure(figsize=(10,6))

all_paths = [] # Lista para guardar todas las trayectorias

for i in range(200):
    steps = np.random.choice([-1,1], size=300)
    path = np.cumsum(steps)
    plt.plot(path, alpha=0.2)
    all_paths.append(path) # Guardar cada trayectoria

# Calcular el promedio de todas las trayectorias
average_path = np.mean(all_paths, axis=0)
plt.plot(average_path, color='black', linewidth=2, label='Promedio de Random Walks') # Graficar el promedio

plt.title("200 Random Walks con Promedio")
plt.legend()
plt.show()
```
