```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def funcion_a_integrar(x):
  return x**2

a = 2
b = 5
N = 100

x = np.random.uniform(a,b,N)
y = np.random.uniform(0,funcion_a_integrar(b),N)
n_bajo_curva = np.sum( y < funcion_a_integrar(x))
respuesta = (b-a) * funcion_a_integrar(b) * n_bajo_curva / N
respuesta
```

```python
plt.plot(x,y, '.')
plt.plot(np.linspace(a,b,100), funcion_a_integrar(np.linspace(a,b,100)), 'r')
plt.show()
```

```python
a = -1
b = 1
N =100

x = np.random.uniform(a,b,N)
y = np.random.uniform(a,b,N)
```

```python
plt.scatter(x,y, label='Random Points')

theta = np.linspace(0, 2*np.pi, 100)
x_circle = np.cos(theta)
y_circle = np.sin(theta)

plt.plot(x_circle, y_circle, 'r-', label='x^2 + y^2 = 1')
plt.legend()
plt.show()
```
